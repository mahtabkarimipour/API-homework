```python
import requests
import json
import random
import csv

url = 'https://api.openweathermap.org/data/2.5/weather'
API_KEY = 'f3f5c70943f5fe3769eaec33b3c91fe2'

with open('data.csv', mode='w') as csv_file:
    fieldnames = ['Name', 'Latitude', 'Temperature', 'WindSpeed', 'Cloudiness', 'Humidity']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    with open('city_list.json', encoding='utf-8') as json_file: 
        l = json.load(json_file)
        random_city_list = random.sample(range(1, len(l)), 500)
        for idx, i in enumerate(random_city_list):
            print('Processing Record', (idx%50) + 1, 'of Set', int(idx/50) + 1, '|', l[i]['name'])
            payload = {'id': l[i]['id'], 'APPID': API_KEY}
            r = requests.get(url, payload)
            print('Request:', r.url)
            info = {}
            info['Name'] = r.json()['name']
            info['Latitude'] = r.json()['coord']['lat']
            info['Temperature'] = r.json()['main']['temp']
            info['WindSpeed'] = r.json()['wind']['speed']
            info['Cloudiness'] = r.json()['clouds']['all']
            info['Humidity'] = r.json()['main']['humidity']
            writer.writerow(info)
```
